```python
def random_player(board):
    move = random.choice(list(board.legal_moves))
    return move
```

```python
def black(board):
    move = random.choice(list(board.legal_moves))
    return move
```

```python
def white(board):
    move = None
    PossibleMoves = list(board.legal_moves)
    IsPossible = False
    while move == None or IsPossible == False:
        move = input('Input your move: ')
        move = chess.Move.from_uci(move)
        if move in PossibleMoves: 
            IsPossible = True
    return move
        
```

```python
def display_board(board,svg):
    if svg:
        return board._repr_svg_()
    else:
        return "<pre>"+str(board)+"</pre>"
```

```python
def checkWinner(board):
    isGameOver = False
    if board.is_checkmate():
            msg = "Checkmate: "
            result = not board.turn
            isGameOver = True
    elif board.is_stalemate():
        msg = "Draw: stalemate"
        isGameOver = True
    elif board.is_insufficient_material():
        msg = "Draw: insufficient material"
        isGameOver = True
    elif board.can_claim_draw():
        msg = "draw:claim"
        isGameOver = True
    print(msg)

    return(msg,board.outcome())
```

```python
def play_game():
    board = chess.Board()
    while not board.is_game_over(claim_draw=True):
        print(board)
        if board.turn == chess.WHITE:
            move = random_player(board)
        if board.turn == chess.BLACK:
            move = black(board)
        board.push(move)    
        print(board.result())
#         eval(board)
        print('----------------------------------------------')
    checkWinner(board)
        
    
```

Let maximising player be black and minimising be white

```python
def minimax(board,depth,isMaximising):
    evaluations = {'1/2-1/2':0, '0-1':-1, '1-0':1}
    result = board.result()
    if(result == '1/2-1/2' or result == '0-1' or result == '1-0'):
        WinScore = evaluations[result]
        return WinScore
    if isMaximizing == True:
        bestScore =-999999999
        possible_moves = list(board.legal_moves)
        for i in range(0,len(possible_moves)):
            board.push(possible_moves[i])
            score = minimax(board,depth+1,False)
            board.pop()
            if score > bestScore:
                bestScore = score 
        return bestScore
    if isMaximizing == False:
        bestScore = 999999999
        possible_moves = list(board.legal_moves)
        for i in range(0,len(possible_moves)):
            board.push(possible_moves[i])
            score = minimax(board,depth+1,True)
            board.pop()
            if score < bestScore:
                bestScore = score
        return bestScore
    
        
```
