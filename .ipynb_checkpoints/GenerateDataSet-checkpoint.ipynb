```python
import chess
import chess.engine
import random
import numpy
```

```python
def random_board(max_depth=200):
    board = chess.Board()
    depth = random.randint(0,max_depth)
    for _ in range(0,depth):
        all_moves = list(board.legal_moves)
        random_move = random.choice(all_moves)
        board.push(random_move)
        if board.is_game_over():
            break
    return board
def stockfish(board,depth):
    with chess.engine.SimpleEngine.popen_uci('./Content/stockfish_15_x64_avx2.exe') as sf:
        result = sf.analyse(board,chess.engine.Limit(depth=depth))
        score = result['score'].white().score()
        return score
```

```python
squares_index = {
    'a':0,
    'b':1,
    'c':2,
    'd':3,
    'e':4,
    'f':5,
    'g':6,
    'h':7
}

def squares_to_index(squares):
    letter = chess.square_name(squares)
    return 8-int(letter[1]),squares_index[letter[0]]

def split_dims(board):
    board3d = numpy.zeros((14,8,8),dtype=numpy.int8)
    
    for piece in chess.PIECE_TYPES:
        for square in board.pieces(piece,chess.WHITE):
            idx= numpy.unravel_index(square,(8,8))
            board3d[piece-1][7-idx[0]][idx[1]] = 1
        for square in board.pieces(piece,chess.BLACK):
            idx = numpy.unravel_index(square,(8,8))
            board3d[piece+5][7-idx[0]][idx[1]] = 1
    aux = board.turn
    board.turn = chess.WHITE
    for move in board.legal_moves:
        i,j = squares_to_index(move.to_square)
        board3d[12][i][j]=1
    board.turn = chess.BLACK
    for move in board.legal_moves:
        i,j = squares_to_index(move.to_square)
        board3d[13][i][j] = 1
    board.turn=aux
    return board3d
```

```python
board = random_board()
print(split_dims(board))
board
```

```python
X,Y = [],[]
```

```python
def GenerateGame():
    score = 0 
    board = random_board()
    score = stockfish(board,10)
    board_serialized = split_dims(board)
    return board,score,board_serialized
```

```python
def GenerateDataSet(NumBoards):
    for i in range(0,NumBoards):
        board,score,serial = GenerateGame()
        if score == None:
            while score == None:
                board, score,serial = GenerateGame()
        print('Generating ... ', i)
        X.append(serial)
        Y.append(score)
```

```python
GenerateDataSet(1000)
X = numpy.array(X)
Y = numpy.array(Y)
print(len(X))
```

```python
numpy.savez('BoardPos',a=X,b=Y)
```
